To install the NetApp ONTAP Python client library, you can use the pip utility. Here are the steps:

1. Make sure you have Python 3.6 or later installed on your system.
2. Install the required packages: `requests` (version 2.26.0 or later), `requests-toolbelt` (version 0.9.1 or later), and `marshmallow` (version 3.2.1 or later)
3. The library requires version 1.26.7 or later of `urllib3` due to outstanding CVEs against older versions. It also requires version 2022.12.7 of `certifi` due to a CVE that removed root certificates from TrustCor from the root store.
